```python
import pandas as pd
```

```python
#sep help to divide column based on ;
df=pd.read_csv("transaction.csv",sep=";")
display(df)
```

### Data cleaning

```python
#summary of dataset
df.info()
```

```python
df['costpertransaction']=df['CostPerItem']*df['NumberOfItemsPurchased']
df
```

```python
df['salespertransaction']=df['SellingPricePerItem']*df['NumberOfItemsPurchased']
```

```python
df['profitpertransaction']=df['salespertransaction']-df['costpertransaction']
```

```python
#markup=(sales-cost)/cost
df['markup']=(df['salespertransaction']-df['costpertransaction'])/df['costpertransaction']
```

```python
df['markup']=round(df['markup'],2)
```

```python
#combining day,month,year
df.dtypes
```

```python
day=df['Day'].astype(str)
year=df['Year'].astype(str)
```

```python
df['date']=day+'-'+df['Month']+'-'+year
df
```

```python
splits=df['ClientKeywords'].str.split(',',expand=True)
splits
```

```python
df['Clientage']=splits[0]
df['ClientType']=splits[1]
df['Length of contarct']=splits[2]
```

```python
df['Clientage']=df['Clientage'].str.replace('[','')
```

```python
df['Length of contarct']=df['Length of contarct'].str.replace(']','')
```

```python
#convert the coulmn to lower
df['ItemDescription'].str.lower()
```

```python
df1=pd.read_csv("value_inc_seasons.csv",sep=";")
```

```python
df1
```

```python
#merge files
df=pd.merge(df,df1,on='Month')
```

```python
df=df.drop('ClientKeywords',axis=1)
```

```python
df=df.drop(['Day','Year','Month'],axis=1)
```

```python
#Export file to csv
df.to_csv("cleaned_project1.csv",index=False)
```
